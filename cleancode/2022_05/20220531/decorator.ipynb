## 다른 유형의 데코레이터  

제너레이터나 코루틴, 이미 데코레이트된 객체도 데코레이트 가능하며, 스택 형태로 쌓을 수 있다.  

## 데코레이터에 인자 전달  

파라메타를 갖는 데코레이터를 구현하는 방법은 여러가지가 있지만, 가장 일반적인 방법은 2가지가 있다.  
1. 간접 참조를 통해 새로운 레벨의 중첩함수를 만들어 데코레이터의 모든 것을 한 단계 더 깊게 만드는 것.  
2. 데코레이터를 위한 클래스를 만드는 것.  

일반적으로는 2번째 방법이 가독성이 더 좋은데 3단계 이상 중첩된 클로저 함수보다는 객체가 더 이해하기 쉽기 때문이다. 

### 중첨 함수의 데코레이터  

데코레이터는 함수를 파라메타로 받아서 함수를 반환하는 함수이다.  

데코레이터를 파라메타에 전달하려면 다른 수준의 간접 참조가 필요한데 1번째 함수는 파라메타를 받아서 내부 함수에 전달하고, 2번째 함수는 데코레이터가 될 함수이며, 3번째는 데코레이팅의 결과를 반환하는 함수이다.
